```python
import os
import numpy as np
import matplotlib.pyplot as plt
import cv2
import pandas as pd
```

```python
train = pd.read_csv("Train/train.csv")
```

```python
images = os.listdir("Train/Images/")
```

```python
len(images)
```

```python
dic= {
    0 : "Pikachu",
    1 : "Bulbasaur",
    2 : "Charmander"
}
reverse_dic= {
    "Pikachu" : 0,
    "Bulbasaur" : 1,
    "Charmander":2
}
```

```python
X= []
y= []
```

```python
for i in images:
    path = "Train/Images/"+i
    img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, (200,200))
    X.append(img)
    lab = reverse_dic[list(train[train['ImageId'] == i]['NameOfPokemon'])[0]]
    y.append(lab)
```

```python
X = np.array(X)
```

```python
y = np.array(y)
```

# kNN

```python
def dist(p1, p2):
    return np.sqrt(np.sum((p1-p2)**2))
def knn(x, y, query_point, k=5):
    m = x.shape[0]
    all_distance = []
    for i in range(m):
        d = dist(query_point, x[i])
        all_distance.append((d, y[i]))
    
    all_distance = sorted(all_distance)[0:k]
    all_distance = np.array(all_distance)[:,1]
    
    unique, freq = np.unique(all_distance, return_counts=True)
    freq = np.argmax(freq)
    return dic[int(unique[freq])]
```

```python
test_images = os.listdir("Test/Images/")
```

```python
for i in test_images[40:50]:
    test_img = cv2.imread("Test/Images/" + i)
    test_img = cv2.cvtColor(test_img, cv2.COLOR_BGR2RGB)
    test_img = cv2.resize(test_img, (200,200))
    plt.imshow(test_img)
    plt.title(knn(X, y, test_img))
    plt.show()
```
