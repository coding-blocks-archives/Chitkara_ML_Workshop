```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
```

```python
x_squared = x**2
```

```python
x_squared
```

## Line Plot

```python
plt.plot(x, x, marker= "*")
plt.plot(x, x_squared, marker="o")
plt.xlabel("Numbers")
plt.ylabel("Comparison")
plt.title("First Graph")
plt.show()
```

## Scatter Graph

```python
X = np.random.randn(1000,2)
```

```python
X.shape
```

```python
x_axis = X[ :  , 0  ]
y_axis = X[ :  , 1 ]
```

```python
plt.scatter(x_axis, y_axis, color="red")
plt.show()
```
